```python
collection_1 = ee.ImageCollection("USGS/NLCD_RELEASES/2019_REL/NLCD")
collection_1
```

```python
map = geemap.Map()
states = ee.FeatureCollection('TIGER/2018/States')
feat = states.filter(ee.Filter.eq('NAME', 'Tennessee'))
image_dem = ee.Image("USGS/3DEP/10m")

image_dem = image_dem.clipToCollection(feat)

vis = {
    'min': 0.0,
    'max': 1500,
    'palette': 'terrain',
}

nlcd_2018 = ee.Image('USGS/NLCD_RELEASES/2019_REL/NLCD/2019').select('landcover').clipToCollection(feat)
left_layer = geemap.ee_tile_layer(nlcd_2018, {}, 'NCLD 2019')
right_layer = geemap.ee_tile_layer(image_dem, vis, 'DEM')

map.centerObject(feat)
map.split_map(left_layer, right_layer, add_close_button=True)
map.add_legend(title="NLCD Land Cover Classification", builtin_legend='NLCD', height='465px')

map.add_colorbar(vis, label="Elevation (m)", layer_name="USGS DEM")
map.addLayer(feat, {}, 'Tennessee')

text = "Made by Andrea Staten"
map.add_text(text, fontsize=20, position='bottomleft')
map
```

```python
map = geemap.Map()
states = ee.FeatureCollection('TIGER/2018/States')
feat = states.filter(ee.Filter.eq('NAME', 'Tennessee'))

collection = (
    ee.ImageCollection("NASA/ORNL/DAYMET_V4")
    .filterDate('2018-08-02', '2018-08-03')
    .select('prcp')
)

image = collection.median().clipToCollection(feat)

vis = {
    'min': 0.0,
    'max': 30,
    'palette': ['1621a2', 'ffffff', '03ffff', '13ff03', 'efff00', 'ffb103', 'ff2300'],
}


nlcd_2018 = ee.Image('USGS/NLCD_RELEASES/2019_REL/NLCD/2019').select('landcover').clipToCollection(feat)
left_layer = geemap.ee_tile_layer(nlcd_2018, {}, 'NCLD 2019')
right_layer = geemap.ee_tile_layer(image, vis, 'Precipitation')

map.centerObject(feat)
map.split_map(left_layer, right_layer, add_close_button=True)
map.add_legend(title="NLCD Land Cover Classification", builtin_legend='NLCD', height='465px')
map.add_colorbar(vis, label="Precipitation Rate (mm)", layer_name="RateBar")
map.addLayer(feat, {}, 'TN Counties')

text = "Made by Andrea Staten"
map.add_text(text, fontsize=20, position='bottomleft')
map
```

```python
nlcd_stats = 'nlcd_stats_sumtn.csv'
geemap.zonal_stats_by_group(
    nlcd_2018,
    feat,
    nlcd_stats,
    statistics_type='SUM',
    denominator=1e6,
    scale=10000,
)
```
